```python
#baca file csv yang ada di folder data
df = pd.read_csv(path + '/data/day.csv')
df
```

```python
#cari jumlah total data
total = len(df)
print("Jumlah total data:", total)
```

```python
#menghitung ukuran tiap pembagian data
data_train =  int(0.8*total) #Data Training: 80% dari total dataset
data_test = total - data_train #Data Testing: 20% dari total dataset
data_val   = int(0.1 * data_train) #Data Validation: 10% dari data training
```

```python
#membagi data data menjadi training, validation, test
train_df = df.iloc[:data_train]
test_df  = df.iloc[data_train:]
val_df = train_df.iloc[:data_val]
train_df = train_df.iloc[data_val:]
```

```python
print("---JUMLAH DATA---\n")
print("Data Training:", len(train_df))
print("Data Validation:", len(val_df))
print("Data Testing:", len(test_df))
```
